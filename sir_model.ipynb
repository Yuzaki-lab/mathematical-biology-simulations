## SIR Model
The SIR model is a classical mathematical model used in epidemiology to describe how infectious diseases spread in a population.
The population is divided into three compartments:

- S : Susceptible (people who can be infected)
- I : Infected (people who are currently infected)
- R : Recoverd (people who recovered and gained immunity)

The total population is assumed  to be constant:

N = S + I + R

## Parameters

β (beta) : infection rate
γ (gamma) : recovery rate

βSI represents the number of new infections per unit time.
γI represents the number of infected individuals who recover per unit time.

## Numerical Simulation
The differential equations cannot always be solved analytically.
Therefore we simulate the system numerically using a time-stepping method.
The population ratios S(t), I(t), and R(t) are updated at each time step.

```python
import numpy as np
import matplotlib.pyplot as plt

beta = 0.3
gamma = 0.1

S = 0.99
I = 0.01
R = 0.0

dt = 0.1
days = 160

S_list = []
I_list = []
R_list = []

for t in range(int(days/dt)):
    dS = -beta * S * I
    dI = beta * S * I - gamma * I
    dR = gamma * I

    S += dS * dt
    I += dI * dt
    R += dR * dt

    S_list.append(S)
    I_list.append(I)
    R_list.append(R)

time = np.arange(0,days,dt)

plt.plot(time,S_list,label="Susceptible")
plt.plot(time,I_list,label="Infected")
plt.plot(time,R_list,label="Recovered")

plt.xlabel("Time")
plt.ylabel("Population ratio")
plt.title("SIR Epidemic Model")
plt.legend()
plt.show()
```

## Simulation Result
The graph shows how the epidemic spreads over time.
Typical behavior:

- The infected population increases rapidly
- It reaches a peak
- Then decreases as people recover

Eventually most individuals move into the recovered group.

## Discussion

The SIR model illustrates how infectious diseases spread through a population by dividing individuals into three compartments: susceptible (S), infectious (I), and recovered (R).

The simulation shows that the number of infected individuals initially increases as susceptible individuals become infected through contact with infectious individuals. After reaching a peak, the infected population begins to decrease as more individuals recover and move into the recovered compartment.

This behavior reflects a typical epidemic curve, where infections rise rapidly at first and then decline as the susceptible population decreases.

Although the SIR model captures the basic mechanism of epidemic spread, it is a simplified representation. Real-world epidemics are influenced by many additional factors such as heterogeneous contact patterns, population structure, vaccination, and behavioral changes, which are not included in the basic model.
