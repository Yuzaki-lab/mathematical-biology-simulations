# Lotka–Volterra Predator–Prey Model Simulation

This notebook simulates the Lotka–Volterra predator–prey model using Python.  
It visualizes both the population dynamics over time and the phase diagram showing interactions between prey and predator populations.

---

## Model

The Lotka–Volterra equations describe interactions between prey (x) and predators (y):

\[
\frac{dx}{dt} = ax - bxy
\]

\[
\frac{dy}{dt} = -cy + dxy
\]

Where:

- **x** : prey population  
- **y** : predator population  
- **a** : prey growth rate  
- **b** : predation rate  
- **c** : predator death rate  
- **d** : predator reproduction rate based on consumed prey

---

## Simulation Method

The differential equations are solved numerically using the **Euler method** with a small time step (`dt = 0.01`) over 50 time units.  
Initial populations:

- Prey: 40  
- Predator: 9

The code below implements the simulation and stores the results for plotting.

### Population Dynamics (b = 0.1)

```python
import numpy as np
import matplotlib.pyplot as plt

a = 1.0
b = 0.1
c = 1.5
d = 0.075

x = 40
y = 9

dt = 0.01
time = np.arange(0,50,dt)

prey = []
predator = []

for t in time:
    dx = a*x - b*x*y
    dy = -c*y + d*x*y

    x = x + dx*dt
    y = y + dy*dt

    prey.append(x)
    predator.append(y)

plt.plot(time, prey, label="prey")
plt.plot(time, predator, label="predator")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.show()
```

### Population Dynamics

This plot shows how prey and predator populations change over time.
The prey population increases first, followed by an increase in the predator population.
This interaction leads to oscillatory dynamics between the two populations.

### Phase Diagram

The phase diagram shows the relationship between the prey and predator populations.

The horizontal axis represents the prey population, while the vertical axis represents the predator population.
The trajectory forms a closed curve, indicating cyclic behavior in the predator–prey system.

### Effect of Parameter b

Comparing the two simulations, increasing the predation rate (b) reduces the amplitude of predator population oscillations.

Because prey are consumed more efficiently, the prey population cannot grow large, which also limits the growth of the predator population.

## Discussion

- The Lotka–Volterra model captures essential predator–prey interactions.

- In reality, additional factors (e.g., environmental limits, multiple species) affect dynamics, which are not included in this simple model.
