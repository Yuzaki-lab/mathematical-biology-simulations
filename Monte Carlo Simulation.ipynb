# Monte Carlo Simulation for Estimating Pi

This notebook demonstrates how the Monte Carlo method can be used to estimate the value of π.

Random points are generated in a unit square, and the proportion that fall inside the quarter circle is used to approximate π.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000

x = np.random.rand(N)
y = np.random.rand(N)

inside = (x**2 + y**2) <= 1

pi_estimate = 4 * np.sum(inside) / N

print("Estimated π:",pi_estimate)

plt.scatter(x[inside], y[inside], s=1)
plt.scatter(x[~inside], y[~inside], s=1)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Monte Carlo Estimation of Pi")
plt.show()
```

# Discussion
As the number of samples increases, the estimation of π becomes more accurate.

This demonstrates how stochastic sampling can be used to approximate numerical values.
