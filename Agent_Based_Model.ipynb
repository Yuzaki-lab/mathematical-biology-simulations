# Agent-Based Model
Even simple rules can generate complex system behavior.

Agent-based modeling is widely used in fields such as economics, biology, and social simulations.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
steps = 100

positions = np.random.rand(N,2)

history = []

for t in range(steps):
    movement = 0.05 * (np.random.rand(N,2) - 0.5)
    positions += movement
    positions = np.clip(positions,0,1)
    history.append(positions.copy())

for pos in history[::10]:
    plt.scatter(pos[:,0],pos[:,1])
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.show()
```
